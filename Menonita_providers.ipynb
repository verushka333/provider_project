```python
menonita_providers['especialidad'] = menonita_providers['especialidad'].str.upper()
```

```python
menonita_providers['especialidad'] = menonita_providers['especialidad'].str.replace('PRÁCTICA', 'PRACTICA').str.replace('PEDIATRÍA','PEDIATRIA').str.replace('GINECÓLOGO','GINECOLOGO').str.replace('GINECOLOGÍA','GINECOLOGIA')
```

```python
menonita_providers['telefono'] = menonita_providers['telefono'].str.replace(r'\(|\)', ' ', regex=True).str.replace(' ', '-').str.replace('-787','787').str.replace('-939', '939').str.replace('-NO-DATA', '').str.replace('--','-')
```

```python
pueblos = menonita_providers['pueblo'].unique()
pueblos.sort()
print(pueblos)
```

```python
menonita_providers['pueblo'] = menonita_providers['pueblo'].str.strip()
```

```python
## Need to add apellido y inicial
```

```python
menonita_providers[['nombre', 'apellidos']] = menonita_providers['nombre'].str.split(' ', n=1, expand=True)
```

```python
def split_apellidos(apellidos):
    if apellidos is None:
        return '', ''
    words = apellidos.split()
    if len(words[0]) == 1:
        return ' '.join(words[1:]), words[0]
    else:
        return apellidos, ''
```

```python
menonita_providers['apellidos'], menonita_providers['inicial'] = zip(*menonita_providers['apellidos'].apply(split_apellidos))
```

```python
menonita_providers['source_id'] = '002'
```

```python
menonita_providers['tipo_plan'] = 'VITAL'
```

```python
columns = ['nombre', 'inicial', 'apellidos', 'especialidad', 'telefono', 'pueblo','tipo_plan','source_id']
menonita_providers = menonita_providers[columns]
```
