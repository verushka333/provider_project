## Cleaning of the "nombre" field

```python
mapfre_providers['apellidos'] = mapfre_providers['apellidos'].str.split('\n').str[0]
```

```python
def split_initial(df, column):
    df['inicial'] = df[column].apply(lambda x: x.strip()[-2:] if x.strip()[-2].isalpha() and x.strip()[-1] == '.' else '')
    df[column] = df[column].apply(lambda x: x.strip()[:-2].strip() if x.strip()[-2].isalpha() and x.strip()[-1] == '.' else x.strip())
    return df

mapfre_providers = split_initial(mapfre_providers, 'apellidos')
```

```python
def split_initial_2(df, column):
    df['inicial'] = df['inicial'] + df[column].apply(lambda x: x.strip().split(' ')[-1] if x[-2] == ' ' else '')
    df[column] = df[column].apply(lambda x: ' '.join(x.strip().split(' ')[:-1]) if x[-2] == ' ' else x)
    return df

mapfre_providers = split_initial_2(mapfre_providers, 'apellidos')
```

```python
def clean_name(name, exclusion_list):
    if any(word in name.upper() for word in exclusion_list):
        return name, '' # Return the current value in 'apellidos' to 'name' and drop it from 'apellidos'
    else:
        return name.strip().split(' ')[-1], ' '.join(name.strip().split(' ')[:-1])

def clean_names(df, column):
    exclusion_list = ['CENTRO', 'GRUPO', 'GROUP', 'CLINICA', 'LAB', 'CENTER', 'HOSPITAL', 'HEALTH', 'CDT', 'SALUD', 'SERVICIOS', 'MUN.','INC','CLINIC','LLC','PUERTO RICO','RADIOLO','MEDICO']
    df['nombre'], df[column] = zip(*df[column].apply(lambda x: clean_name(x, exclusion_list)))
    return df

mapfre_providers = clean_names(mapfre_providers, 'apellidos')
```

```python
def split_del(df):
    df.loc[df['nombre'].str.contains('DEL'), 'nombre'] = df.apply(lambda row: row['apellidos'].split(' ')[0] + ' ' + row['nombre'] if 'DEL' in row['nombre'] else row['nombre'], axis=1)
    df.loc[df['nombre'].str.contains('DEL'), 'apellidos'] = df['apellidos'].apply(lambda x: ' '.join(x.split(' ')[1:]) if 'DEL' in x else x)
    df.loc[df['nombre'].str.contains('DEL'), 'nombre'] = df['nombre'].apply(lambda x: ' '.join(x.split(' ')[::-1]) if 'DEL' in x else x)
    return df

mapfre_providers = split_del(mapfre_providers)
```

### New "categoria" field

```python
def assign_categoria(especialidad):
    if especialidad.startswith('LAB'):
        return 'LABORATORIOS'
    elif especialidad.startswith('HOSPITAL') or especialidad == 'SALA DE EMERGENCIA':
        return 'HOSPITAL'
    elif 'PSIQUI' in especialidad or 'PSICO' in especialidad:
        return 'SALUD MENTAL'
    elif especialidad == 'CENTRO DE RADIOLOGICO':
        return 'RAYOS X'
    elif 'SALUD MENTAL' in especialidad:
        return 'SALUD MENTAL'
    elif 'PSIQUI' in especialidad:
        return 'SALUD MENTAL'
    elif 'CENTRO' in especialidad or 'CENTER' in especialidad:
        return 'CENTROS DE SALUD'
    elif especialidad == 'CENTRO MULTIESPECIALIDADES':
        return 'MEDICOS'
    elif 'DENT' in especialidad or 'MAXI' in especialidad or 'ORAL' in especialidad or 'ENDODO' in especialidad:
        return 'DENTAL'
    elif 'PEDODON' in especialidad or 'ORTODONCISTA' in especialidad or 'PROSTHO' in especialidad or 'PERIODON' in especialidad:
        return 'DENTAL'
    elif especialidad.startswith('FACILIDAD') or especialidad.startswith('CLINICA'):
        return ''
    else:
        return 'MEDICOS'
```

```python
mapfre_providers['especialidad']= mapfre_providers['especialidad'].astype(str)
```

```python
mapfre_providers['categoria'] = mapfre_providers['especialidad'].apply(assign_categoria)
```

## Cleaning of the "direccion" field

```python
def process_direccion(df, column):
    df['zipcode'] = df[column].apply(lambda x: x.split()[-1])
    df['estado'] = df[column].apply(lambda x: x.split()[-2][-2:])
    df['pueblo'] = df[column].apply(lambda x: x.split(',PR')[0].split('\n')[-1].strip() if ',PR' in x else None)
    df['direccion_2'] = df[column].apply(lambda x: x.split('\n')[-2].split(',')[1] if ',' in x.split('\n')[-2] else x.split('\n')[-2])
    df[column] = df[column].apply(lambda x: x.split('\n')[0])
    return df

mapfre_providers = process_direccion(mapfre_providers, 'direccion')
```

```python
mapfre_providers['pueblo'] = mapfre_providers['pueblo'].astype(str)
```

```python
def loose_pueblos(pueblo):
    if pueblo== 'COTO LAUREL':
        return 'PONCE'
    elif pueblo == 'ISLA VERDE':
        return 'CAROLINA'
    elif pueblo == 'HATO REY':
        return 'SAN JUAN'
    elif pueblo == 'SANTURCE':
        return 'SAN JUAN'
    elif pueblo == 'RIO PIEDRAS':
        return 'SAN JUAN'
    elif pueblo == 'SABANA SECA':
        return 'TOA BAJA'
    elif pueblo == 'MERCEDITA':
        return 'PONCE'
    elif pueblo == 'CASTANER':
        return 'LARES'
    else:
        return pueblo
```

```python
mapfre_providers['pueblo'] = mapfre_providers['pueblo'].apply(loose_pueblos)
```

```python
mapfre_providers['pueblo'] = mapfre_providers['pueblo'].str.replace('Ponce','PONCE').str.replace('Ñ','N').str.replace('Caguas','CAGUAS').str.replace('carolina','CAROLINA')
```
