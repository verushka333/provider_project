## Creating DataFrame

```python
import pandas as pd
```

```python
data = {
    'name': nombre_values,
    'especialidad': especialidad_values,
    'telefono': telefono_values,
    'direccion': direccion_values
}
mapfre_providers = pd.DataFrame(data)
```

```python
mapfre_providers
```

## Cleaning of the "nombre" field

```python
mapfre_providers['name'] = mapfre_providers['name'].str.split('\n').str[0]
```

```python
def split_initial(df, column):
    df['inicial'] = df[column].apply(lambda x: x.strip()[-2:] if x.strip()[-2].isalpha() and x.strip()[-1] == '.' else '')
    df[column] = df[column].apply(lambda x: x.strip()[:-2].strip() if x.strip()[-2].isalpha() and x.strip()[-1] == '.' else x.strip())
    return df

mapfre_providers = split_initial(mapfre_providers, 'name')
```

```python
mapfre_providers['nombre'] = mapfre_providers['name'].apply(lambda x: x.split()[-1])
mapfre_providers['name'] = mapfre_providers['name'].apply(lambda x: ' '.join(x.split()[:-1]))
```

```python
mapfre_providers = mapfre_providers.rename(columns={'name': 'apellidos'})
```

```python
mapfre_providers
```

## Cleaning of the "especialidad" field

```python
mapfre_providers['especialidad'].unique()
```

```python
mapfre_providers['especialidad'] = mapfre_providers['especialidad'].str.replace('&', ' ').str.replace('  ',' ')
```

```python
mapfre_providers['especialidad'] = mapfre_providers['especialidad'].str.replace('ENFERMEDADES INFECCIOSAS', 'INFECTOLOGIA').str.replace('PATOLOGIA DE HABLA', 'PATOLOGIA DEL HABLA').str.replace('MEDICINA GERIATRICA', 'GERIATRIA')
mapfre_providers['especialidad'] = mapfre_providers['especialidad'].str.replace('MEDICINA GENERAL','GENERALISTA').str.replace('CIRUGIA DE COLON Y RECTO', 'CIRUGIA COLON RECTAL').str.replace('MEDICINA FISICA PEDIATRICA', 'FISIATRIA PEDIATRICA').str.replace('CIRUGIA TORACICA', 'CIRUGIA TORAXICA')
```

```python
mapfre_providers['especialidad'] = mapfre_providers['especialidad'].str.replace('  ',' ').str.replace('/','')
```

### New "categoria" field

```python
def assign_categoria(especialidad):
    if especialidad.startswith('LAB'):
        return 'LABORATORIOS'
    elif especialidad.startswith('HOSPITAL') or especialidad == 'SALA DE EMERGENCIA' or 'PSIQUI' in especialidad:
        return 'HOSPITAL'
    elif especialidad == 'CENTRO DE RADIOLOGICO':
        return 'RAYOS X'
    elif 'SALUD MENTAL' in especialidad:
        return 'SALUD MENTAL'
    elif 'PSIQUI' in especialidad:
        return 'SALUD MENTAL'
    elif especialidad.startswith('CENTRO'):
        return 'CENTROS DE SALUD'
    elif especialidad == 'CENTRO MULTIESPECIALIDADES':
        return 'MEDICOS'
    elif especialidad.startswith('FACILIDAD') or especialidad.startswith('CLINICA'):
        return ''
    else:
        return 'MEDICOS'
```

```python
mapfre_providers['especialidad']= mapfre_providers['especialidad'].astype(str)
```

```python
mapfre_providers['categoria'] = mapfre_providers['especialidad'].apply(assign_categoria)
```

```python
mapfre_providers['categoria'].value_counts()
```

```python
mapfre_providers
```

## Cleaning of the "direccion" field

```python
def process_direccion(df, column):
    df['zipcode'] = df[column].apply(lambda x: x.split()[-1])
    df['estado'] = df[column].apply(lambda x: x.split()[-2][-2:])
    df['pueblo'] = df[column].apply(lambda x: x.split(',PR')[0].split('\n')[-1].strip() if ',PR' in x else None)
    df['direccion_2'] = df[column].apply(lambda x: x.split('\n')[-2].split(',')[1] if ',' in x.split('\n')[-2] else x.split('\n')[-2])
    df[column] = df[column].apply(lambda x: x.split('\n')[0])
    return df

mapfre_providers = process_direccion(mapfre_providers, 'direccion')
```

```python
mapfre_providers['estado'].value_counts()
```

```python
mapfre_providers['pueblo'].nunique()
```

```python
mapfre_providers['pueblo'] = mapfre_providers['pueblo'].astype(str)
```

```python
def loose_pueblos(pueblo):
    if pueblo== 'COTO LAUREL':
        return 'PONCE'
    elif pueblo == 'ISLA VERDE':
        return 'CAROLINA'
    elif pueblo == 'HATO REY':
        return 'SAN JUAN'
    elif pueblo == 'SANTURCE':
        return 'SAN JUAN'
    elif pueblo == 'RIO PIEDRAS':
        return 'SAN JUAN'
    elif pueblo == 'SABANA SECA':
        return 'TOA BAJA'
    elif pueblo == 'MERCEDITA':
        return 'PONCE'
    elif pueblo == 'CASTANER':
        return 'LARES'
    else:
        return pueblo
```

```python
mapfre_providers['pueblo'] = mapfre_providers['pueblo'].apply(loose_pueblos)
```

```python
mapfre_providers['pueblo'] = mapfre_providers['pueblo'].str.replace('Ponce','PONCE').str.replace('Ñ','N').str.replace('Caguas','CAGUAS').str.replace('carolina','CAROLINA')
```

```python
pueblos = mapfre_providers['pueblo'].unique()
pueblos.sort()
print(pueblos)
```
