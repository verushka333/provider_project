## Creating DataFrame

```python
import pandas as pd
```

### Loop that scraps from page 1 to page 1003

```python
from selenium.common.exceptions import TimeoutException

# Initialize an empty DataFrame
mcs = pd.DataFrame()

# Set the total number of pages
num_of_clicks = 1003
wait = WebDriverWait(driver, 80) 

# Get the current window handle (the original window)
original_window = driver.current_window_handle

# Loop through each page
for i in range(1, num_of_clicks + 1):
    wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, '.table>tbody>tr')))

    # Find all the rows on the current page
    rows = driver.find_elements(By.CSS_SELECTOR, '.table>tbody>tr')

    # Loop through each row
    for row in rows:
        cells = row.find_elements(By.TAG_NAME, 'td')
        
        # Extract the text from each cell
        nombre = cells[0].text
        categoria = cells[1].text
        especialidad = cells[2].text
        telefono = cells[3].text
        pueblo = cells[4].text

        # Create a DataFrame for the current record
        record_df = pd.DataFrame({
            'nombre': [nombre],
            'categoria': [categoria],
            'especialidad': [especialidad],
            'telefono': [telefono],
            'pueblo': [pueblo]
        })

        # Concatenate the current record DataFrame to the main DataFrame
        mcs = pd.concat([mcs, record_df], ignore_index=True)
        
    # Click to go to the next page
    try:
        next_page_element = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, f"a.change-page[data-page='{i + 1}']")))
        next_page_element.click()
        print("Next button clicked")  # Check if the "next" button is being clicked
        time.sleep(5)  # Wait for the page to refresh
        print(driver.current_url)  # Check if the page URL is changing
    except StaleElementReferenceException:
        continue
    except TimeoutException:
        print("Reached the last page.")
        break  # Break the loop if the 'Next' button is not found (i.e., we're on the last page)


driver.quit()

print(mcs)
```

## Cleaning of the "nombre" field

```python
def split_initial_end(df, column):
    df['inicial'] = df[column].apply(lambda x: x[-2:] if x[-1].isalpha() and x[-2] == ' ' else '')
    df[column] = df[column].apply(lambda x: x[:-2].strip() if x[-1].isalpha() and x[-2] == ' ' else x.strip())
    return df
```

```python
mcs_limited = split_initial_end(mcs_limited, 'nombre')
```

```python
mcs_limited['inicial'].value_counts()
```

```python
mcs_limited
```

```python
def split_name(df, column):
    df['name'] = df[column].apply(lambda x: ' '.join(x.split()[-2:]) if x.split()[-1] == 'DEL' else x.split()[-1])
    df[column] = df[column].apply(lambda x: ' '.join(x.split()[:-2]) if x.split()[-1] == 'DEL' else ' '.join(x.split()[:-1]))
    return df
```

```python
mcs_limited = split_name(mcs_limited, 'nombre')
```

## Cleaning of the "telefono" field

```python
def format_number(number):
    # Remove any existing dashes
    str_num = str(number).replace('-', '')
    
    # Check if the number has 10 digits
    if len(str_num) == 10:
        formatted_num = "{}-{}-{}".format(str_num[:3], str_num[3:6], str_num[6:])
    else:
        formatted_num = number  # If not, return the original number

    return formatted_num

print(format_number(7878743760))  # Output: 787-874-3760
```

```python
mcs_limited['telefono'] =mcs_limited['telefono'].apply(format_number)
```

```python
mcs_limited['telefono'] = mcs_limited['telefono'].apply(lambda x: x.split(' EXT ')[0])
```
