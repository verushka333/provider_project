## Cleaning of the "direccion" fields

```python
firstmedical['direccion'] = firstmedical['direccion'].astype(str)
```

```python
def split_address_estado(address):
    words = address.split()
    if len(words) < 2:
        return address, '', ''
    else:
        return ' '.join(words[:-2]), ' '.join(words[-2:])

firstmedical['direccion'], firstmedical['estado'] = zip(*firstmedical['direccion'].apply(split_address_estado))
```

```python
firstmedical['estado']=firstmedical['estado'].str.replace('PUERTO RICO','PR')
```

```python
firstmedical['estado'] = firstmedical['estado'].str.replace('CALIFORNIA CALIFORNIA','CA')
```

```python
def split_address_pueblo(address):
    exception_words = ['BUENAS', 'ROJO', 'DIAZ', 'MARIAS', 'PIEDRAS', 'GRANDE', 'GERMAN', 'JUAN', 'ALTA', 'BAJA', 'ALTO', 'SEBASTIAN', 'LORENZO', 'VERDE', 'LAUREL', 'SECA', 'REY', 'ISABEL']
    words = address.split()
    if len(words) < 2:
        return address, '', ''
    else:
        if words[-1] in exception_words:
            return ' '.join(words[:-2]), ' '.join(words[-2:])
        else:
            return ' '.join(words[:-1]), ' '.join(words[-1:])

firstmedical['direccion'], firstmedical['pueblo'] = zip(*firstmedical['direccion'].apply(split_address_pueblo))
```

```python
firstmedical['pueblo'] = firstmedical['pueblo'].str.replace('MAYAGÜEZ','MAYAGUEZ')
firstmedical['pueblo'] = firstmedical['pueblo'].str.replace('CASTANER','LARES')
```

```python
firstmedical = firstmedical[firstmedical['pueblo'] != 'DR']
```

```python
pueblos = firstmedical['pueblo'].unique()
pueblos.sort()
print(pueblos)
```

```python
def loose_pueblos(pueblo):
    if pueblo== 'COTO LAUREL':
        return 'PONCE'
    elif pueblo == 'ISLA VERDE':
        return 'CAROLINA'
    elif pueblo == 'HATO REY':
        return 'SAN JUAN'
    elif pueblo == 'SANTURCE':
        return 'SAN JUAN'
    elif pueblo == 'RIO PIEDRAS':
        return 'SAN JUAN'
    elif pueblo == 'SABANA SECA':
        return 'TOA BAJA'
    else:
        return pueblo
```

```python
firstmedical['pueblo'] = firstmedical['pueblo'].apply(loose_pueblos)
```

## Cleaning of the "especialidad" fields

```python
def assign_categoria(especialidad):
    if especialidad.startswith('LAB'):
        return 'LABORATORIOS'
    elif especialidad.startswith('HOSPITAL') or especialidad == 'SALA DE EMERGENCIA':
        return 'HOSPITAL'
    elif especialidad == 'CENTRO DE RADIOLOGIA':
        return 'RAYOS X'
    elif 'SALUD MENTAL' in especialidad:
        return 'SALUD MENTAL'
    elif 'PSIQUI' in especialidad:
        return 'SALUD MENTAL'
    elif especialidad.startswith('CENTRO') and especialidad != 'CENTRO DE RADIOLOGIA':
        return ''
    else:
        return 'MEDICOS'
    
```

```python
firstmedical['especialidad'] = firstmedical['especialidad'].astype(str)
```

```python
firstmedical['categoria'] = firstmedical['especialidad'].apply(assign_categoria)
```

```python
firstmedical['especialidad'] = firstmedical['especialidad'].str.replace('OBSTETRICIA Y GINECOLOGIA','GINECOLOGIA OBSTETRICIA').str.replace(' E ',' ').str.replace('Y','').str.replace('  ',' ')
```

```python
firstmedical['especialidad'] = firstmedical['especialidad'].str.replace('DENTISTA ENDODONCIA','ENDODONCIA').str.replace('MEDICINA GENERAL','GENERALISTA')
```

## Cleaning of the "nombre" fields

```python
# Create a DataFrame for rows where 'Nombre' starts with 'LAB', 'CENTRO', 'HOSPITAL', 'CDT', 'PUERTO', 'CONSOLIDATED', or 'SALUD'
nomed_rows = firstmedical[firstmedical['nombre'].str.startswith('LAB') | firstmedical['nombre'].str.startswith('CENTRO') | firstmedical['nombre'].str.startswith('HOSPITAL') | firstmedical['nombre'].str.startswith('CDT') | firstmedical['nombre'].str.startswith('PUERTO') | firstmedical['nombre'].str.startswith('CONSOLIDATED') | firstmedical['nombre'].str.startswith('SALUD')]

# Create a DataFrame for the rest of the rows
other_rows = firstmedical[~(firstmedical['nombre'].str.startswith('LAB') | firstmedical['nombre'].str.startswith('CENTRO') | firstmedical['nombre'].str.startswith('HOSPITAL') | firstmedical['nombre'].str.startswith('CDT') | firstmedical['nombre'].str.startswith('PUERTO') | firstmedical['nombre'].str.startswith('CONSOLIDATED') | firstmedical['nombre'].str.startswith('SALUD'))]

# Apply the split function to 'other_rows'
other_rows[['nombre', 'apellidos']] = other_rows['nombre'].str.split(' ', n=1, expand=True)

# Concatenate 'lab_rows' and 'other_rows'
firstmedical = pd.concat([nomed_rows, other_rows])
```

```python
firstmedical['apellidos'] = firstmedical['apellidos'].astype(str)
```

```python
def split_apellidos(apellidos):
    if apellidos is None:
        return '', ''
    words = apellidos.split()
    if len(words[0]) == 1:
        return ' '.join(words[1:]), words[0]
    else:
        return apellidos, ''

firstmedical['apellidos'], firstmedical['inicial'] = zip(*firstmedical['apellidos'].apply(split_apellidos))
```
