```python
firstmedical['Teléfono #1'] = firstmedical['Teléfono #1'].str.replace(r'\(|\)', ' ', regex=True).str.replace(' ', '-').str.replace('-787','787').str.replace('-939', '939')
```

```python
firstmedical['Teléfono #2'] = firstmedical['Teléfono #2'].str.replace(r'\(|\)', ' ', regex=True).str.replace(' ', '-').str.replace('-787','787').str.replace('-939', '939')
```

```python
firstmedical['Teléfono #1'] = firstmedical['Teléfono #1'].str.replace('--','-')
firstmedical['Teléfono #2'] = firstmedical['Teléfono #2'].str.replace('--','-')
```

```python
# Fill NaN values with an empty string
firstmedical = firstmedical.fillna('')

# Create a DataFrame for rows where 'Nombre' starts with 'LAB'
lab_rows = firstmedical[firstmedical['Nombre'].str.startswith('LAB')]

# Create a DataFrame for the rest of the rows
other_rows = firstmedical[~firstmedical['Nombre'].str.startswith('LAB')]

# Apply the split function to 'other_rows'
other_rows[['Nombre', 'Apellidos']] = other_rows['Nombre'].str.split(' ', n=1, expand=True)

# Concatenate 'lab_rows' and 'other_rows'
firstmedical = pd.concat([lab_rows, other_rows])
```

```python
firstmedical['Dirección'] = firstmedical['Dirección'].astype(str)
```

```python
firstmedical
```

```python
def split_address(address):
    words = address.split()
    if len(words) < 2:
        return address, '', ''
    else:
        return ' '.join(words[:-2]), ' '.join(words[-2:])

firstmedical['Dirección'], firstmedical['estado'] = zip(*firstmedical['Dirección'].apply(split_address))
```

```python
firstmedical
```

```python
def split_address_pueblo(address):
    exception_words = ['BUENAS', 'ROJO', 'DIAZ', 'MARIAS', 'PIEDRAS', 'GRANDE', 'GERMAN', 'JUAN', 'ALTA', 'BAJA', 'ALTO', 'SEBASTIAN', 'LORENZO', 'VERDE', 'LAUREL', 'SECA', 'REY', 'ISABEL']
    words = address.split()
    if len(words) < 2:
        return address, '', ''
    else:
        if words[-1] in exception_words:
            return ' '.join(words[:-2]), ' '.join(words[-2:])
        else:
            return ' '.join(words[:-1]), ' '.join(words[-1:])

firstmedical['Dirección'], firstmedical['pueblo'] = zip(*firstmedical['Dirección'].apply(split_address_pueblo))
```

```python
def split_apellidos(apellidos):
    if apellidos is None:
        return '', ''
    words = apellidos.split()
    if len(words[0]) == 1:
        return ' '.join(words[1:]), words[0]
    else:
        return apellidos, ''

firstmedical['apellidos'], firstmedical['inicial'] = zip(*firstmedical['apellidos'].apply(split_apellidos))
```
