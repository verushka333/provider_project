```python
## This code is working and loads information to database

url = "https://service.ssspr.com/salud/provider/search/"
browser=webdriver.Chrome()
response = browser.get(url)

# Initialize an empty DataFrame
TripleS = pd.DataFrame()
current_page = 1
wait = WebDriverWait(browser, 10)
directory = ["PPO","PPOELA","PPOFEDBCBS","PPOFED"]
#pueblo_value = "Moca                                                        "   

# Get the current window handle (the original window)
original_window = browser.current_window_handle

# This value selection is working
for d in directory:
    # Refresh the page before starting the next iteration
    browser.refresh()
    WebDriverWait(browser, 10).until(EC.presence_of_element_located((By.CSS_SELECTOR, '#ctl00_Content_ddlDirectory')))

    directory_dropdown = Select(browser.find_element(By.CSS_SELECTOR, '#ctl00_Content_ddlDirectory'))
    directory_dropdown.select_by_value(d)
    
    print(f"selected directory {d}")
    
    for p in pueblo:
        current_page = 1
        browser.refresh()
        WebDriverWait(browser, 10).until(EC.visibility_of_element_located((By.XPATH, '//*[@id="ctl00_Content_ddlTown"]')))

        pueblo_dropdown = Select(browser.find_element('xpath', '//*[@id="ctl00_Content_ddlTown"]'))
        pueblo_dropdown.select_by_value(p)
        
        print(f"selected town {p}")
        
        #Click the "Buscar" button
        wait = WebDriverWait(browser, 10)
        element = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="ctl00_Content_search"]')))
        element.click()
        
        
        while True:
            try:
                # Wait for the page number to be visible
                page_link = WebDriverWait(browser, 20).until(EC.presence_of_element_located((By.LINK_TEXT, str(current_page))))

                # Scroll to the element before clicking it
                browser.execute_script("arguments[0].scrollIntoView();", page_link)

                # Use JavaScript to click the element
                browser.execute_script("arguments[0].click();", page_link)

                print(f"{current_page} little piggy")
                time.sleep(5)  # Wait for the page to refresh
                
                # Fetch elements from the browser
                npi_find = browser.find_elements('css selector', '[data-provider]')
                npi = [element.get_attribute('data-provider') for element in npi_find]                
                apellidos = browser.find_elements('css selector', 'li.large-2.columns.name')[1:]
                direccion = browser.find_elements('css selector', 'li.large-2.columns.address')[1:]
                pueblos = browser.find_elements('css selector', 'li.large-1.columns.cityname')[1:]
                especialidad = browser.find_elements('css selector','li.large-2.columns.speciality')[1:]
                telefono = browser.find_elements('css selector', 'li.large-2.columns.phone')[1:]
                estrellas_find = browser.find_elements('css selector', 'div.estrellas > [alt]')
                estrellas = [element.get_attribute('alt') for element in estrellas_find]

                
                # Update the TripeS DataFrame
                for j in range(len(apellidos)):
                    data = pd.DataFrame({
                        'npi': [npi[j]],
                        'apellidos': [apellidos[j].text],
                        'direccion': [direccion[j].text],
                        'pueblo': [pueblos[j].text],
                        'especialidad':[especialidad[j].text],
                        'telefono': [telefono[j].text],
                        'estrellas': [estrellas[j] if j < len(estrellas) else ''],
                        
                    })
                
                TripleS = pd.concat([TripleS, data], ignore_index=True)
            
                current_page += 1
            
            except NoSuchElementException:
                print("No additional pages found. Continuing...")
                break

            except StaleElementReferenceException:
                continue

            except TimeoutException:
                print("Reached the last page.")
                break  # Break the loop if the 'Next' button is not found (i.e., we're on the last page)
            
print(TripleS)        
            
        

```

```python
def split_initial_end(df, column):
    df['inicial'] = df[column].apply(lambda x: x[-2:] if x[-1].isalpha() and x[-2] == ' ' else '')
    df[column] = df[column].apply(lambda x: x[:-2].strip() if x[-1].isalpha() and x[-2] == ' ' else x.strip())
    return df
```

```python
def split_name(df, column):
    df['nombre'] = df[column].apply(lambda x: ' '.join(x.split()[-2:]) if x.split()[-1] == 'DEL' else x.split()[-1])
    df[column] = df[column].apply(lambda x: ' '.join(x.split()[:-2]) if x.split()[-1] == 'DEL' else ' '.join(x.split()[:-1]))

    # Add the new condition
    conditions = ['LAB', 'CENTRO', 'HOSPITAL', 'CDT', 'PUERTO', 'CONSOLIDATED', 'INC', 'SALUD']
    for cond in conditions:
        mask = df['nombre'].str.startswith(cond)
        df.loc[mask, 'nombre'] = df.loc[mask, 'apellidos']
        df.loc[mask, 'apellidos'] = ''

    return df
```

## New "categoria" field

```python
def assign_categoria(especialidad):
    if especialidad.startswith('LAB'):
        return 'LABORATORIOS'
    elif especialidad.startswith('HOSPITAL') or especialidad == 'SALA DE EMERGENCIA':
        return 'HOSPITAL'
    elif 'PSIQUI' in especialidad or 'PSICO' in especialidad:
        return 'SALUD MENTAL'
    elif especialidad == 'CENTRO DE RADIOLOGICO':
        return 'RAYOS X'
    elif 'SALUD MENTAL' in especialidad:
        return 'SALUD MENTAL'
    elif 'PSIQUI' in especialidad:
        return 'SALUD MENTAL'
    elif 'LAB' in especialidad:
        return 'LABORATORIOS'
    elif 'CENTRO' in especialidad or 'CENTER' in especialidad:
        return 'CENTROS DE SALUD'
    elif especialidad == 'CENTRO MULTIESPECIALIDADES':
        return 'MEDICOS'
    elif 'DENT' in especialidad or 'MAXI' in especialidad or 'ORAL' in especialidad or 'ENDODO' in especialidad:
        return 'DENTAL'
    elif 'PEDODON' in especialidad or 'ORTODONCISTA' in especialidad or 'PROSTHO' in especialidad or 'PERIODON' in especialidad:
        return 'DENTAL'
    elif especialidad.startswith('FACILIDAD') or especialidad.startswith('CLINICA'):
        return ''
    else:
        return 'MEDICOS'
```

```python
TripleS_copy['categoria'] = TripleS_copy['especialidad'].apply(assign_categoria)
```

## Cleaning of the "telefono" & "email" field

```python
# Initialize new columns for the additional phone numbers
TripleS_copy['email'] = ""
TripleS_copy['email_alt'] = ""

# Iterate over the DataFrame rows
for i, row in TripleS_copy.iterrows():
    # Create new lists for the 'telefono' and 'email' columns
    new_phone_list = []
    new_email_list = []
    
    # Split the 'telefono' field into a list of values using '\n' as the separator
    telefono_list = row['telefono'].split('\n')
    
    # Iterate over each value in the list
    for value in telefono_list:
        # If the value contains '@', it's an email
        if '@' in value:
            # Add it to the 'email' column
            new_email_list.append(value)
        else:
            # Otherwise, add it to the 'telefono' column
            new_phone_list.append(value)
    
    # Assign the new lists back to the DataFrame
    TripleS_copy.at[i, 'telefono'] = ' & '.join(new_phone_list)
    
    # If there are more than one email
    if len(new_email_list) > 1:
        # Assign the second email to 'email_alt'
        TripleS_copy.at[i, 'email_alt'] = new_email_list[1]
        
    # Always keep the first email in 'email'
    TripleS_copy.at[i, 'email'] = new_email_list[0] if new_email_list else ""
```

```python
# Initialize new columns for the additional phone numbers
TripleS_copy['telefono_1'] = ''
TripleS_copy['telefono_2'] = ''

# Iterate over the DataFrame rows
for i, row in TripleS_copy.iterrows():
    # Split the 'telefono' field into a list of phone numbers
    phone_numbers = row['telefono'].split(' & ')
    
    # Always keep the first phone number in 'telefono'
    TripleS_copy.at[i, 'telefono'] = phone_numbers[0]
    
    # If there are more than one phone number
    if len(phone_numbers) > 1:
        # Assign the second phone number to 'telefono_1'
        TripleS_copy.at[i, 'telefono_1'] = phone_numbers[1]
        
    # If there are more than two phone numbers
    if len(phone_numbers) > 2:
        # Assign the third phone number to 'telefono_2'
        TripleS_copy.at[i, 'telefono_2'] = phone_numbers[2]
        
```

```python
# Initialize new columns for the additional phone numbers
TripleS_copy['email_alt'] = ''

# Iterate over the DataFrame rows
for i, row in TripleS_copy.iterrows():
    # Split the 'telefono' field into a list of phone numbers
    email_tog = row['email'].split(' & ')
    
    # Always keep the first phone number in 'telefono'
    TripleS_copy.at[i, 'email'] = email_tog[0]
    
    # If there are more than one phone number
    if len(email_tog) > 1:
        # Assign the second phone number to 'telefono_1'
        TripleS_copy.at[i, 'email_alt'] = email_tog[1]
        
    # If 'email' is equal to 'email_alt'
    if TripleS_copy.at[i, 'email'] == TripleS_copy.at[i, 'email_alt']:
        # Set the value of 'email_alt' to None
        TripleS_copy.at[i, 'email_alt'] = ''
        
```

```python
def format_phone_number(phone_number):
    # Remove existing dashes if any
    phone_number = phone_number.replace('-', '')

    # Group into sets of three for the first two groups and four for the last group
    formatted_phone_number = '-'.join([phone_number[i:i+3] for i in range(0, 6, 3)])
    formatted_phone_number += '-' + phone_number[6:]  # Append the last four digits

    return formatted_phone_number
```

```python
TripleS_copy['telefono'] = TripleS_copy['telefono'].apply(format_phone_number)
```

```python
TripleS_copy['telefono_1'] = TripleS_copy['telefono_1'].apply(format_phone_number)
```

```python
TripleS_copy['telefono_2'] = TripleS_copy['telefono_2'].apply(format_phone_number)
```

```python
def remove_duplicate_phones(row):
    telefono, telefono_1, telefono_2 = row['telefono'], row['telefono_1'], row['telefono_2']
    if telefono == telefono_1:
        telefono_1 = ""
    if telefono_1 == telefono_2:
        telefono_2 = ""
    return pd.Series([telefono, telefono_1, telefono_2])
```

```python
TripleS_copy[['telefono', 'telefono_1', 'telefono_2']] = TripleS_copy.apply(remove_duplicate_phones, axis=1)
```
