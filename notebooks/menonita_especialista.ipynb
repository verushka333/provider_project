## Creating DataFrame

```python
import pandas as pd
```

```python
menonita_especialistas
```

## Cleaning of the "especialidad" field

```python
menonita_especialistas['especialidad'].value_counts()
```

```python
menonita_especialistas['especialidad'] = menonita_especialistas['especialidad'].str.upper()
```

```python
menonita_especialistas['especialidad'] = menonita_especialistas['especialidad'].str.replace('PRCTICA GENERAL', 'GENERALISTA').str.replace('PEDIATRA','PEDIATRIA').str.replace('GINECLOGO OBSTETRA','GINECOLOGIA OBSTETRICIA').str.replace('GINECOLOGA','GINECOLOGO').str.replace('GERIATRA','GERIATRIA').str.replace('MEDICINA FAMILIAR','MEDICINA DE FAMILIA')
```

```python
menonita_especialistas['especialidad'] = menonita_especialistas['especialidad'].str.replace('', 'A').str.replace('','I').str.replace('','O').str.replace(' E ',' ').str.replace('','E').str.replace('','U')
```

```python
menonita_especialistas['especialidad'] = menonita_especialistas['especialidad'].str.replace('CUIDADO CRITICO', 'CUIDADO INTENSIVO').str.replace('DENTISTA GENERAL','DENTISTA').str.replace('ANESTESIA','ANESTESIOLOGIA').str.replace('CARDIOLOGO PEDIATRICO','CARDIOLOGIA PEDIATRICA').str.replace('CIRUGIA COLORECTAL','CIRUGIA COLON RECTAL')
menonita_especialistas['especialidad'] = menonita_especialistas['especialidad'].str.replace('CIRUGIA TORACICA CARDIOVASCULAR', 'CIRUGIA TORAXICA CARDIOVASCULAR').str.replace('ENDOCRINOLOGO','ENDOCRINOLOGIA').str.replace('ONCOLOGO','ONCOLOGIA').str.replace('ONCOLOGO PEDIATRICO','ONCOLOGIA PEDIATRICA')
menonita_especialistas['especialidad'] = menonita_especialistas['especialidad'].str.replace('OTORRALINGOLOGO PEDIATRICO', 'OTORRINOLARINGOLOGIA PEDIATRICA').str.replace('PATOLOGA DEL HABLA','PATOLOGIA DEL HABLA').str.replace('PODIATRA','PODIATRIA').str.replace('NEUMOLOGO PEDIATRICO','NEUMOLOGIA PEDIATRICA')
```

```python
menonita_especialistas['categoria'] = 'MEDICOS'
```

```python
u = pd.Series(menonita_especialistas['especialidad'].unique())
print(u.head(60))
```

## Cleaning of the "telefono" field

```python
menonita_especialistas['telefono'] =menonita_especialistas['telefono'].str.replace(r'\(|\)', ' ', regex=True).str.replace(' ', '-').str.replace('-787','787').str.replace('-939', '939').str.replace('-NO-DATA', '').str.replace('--','-')
```

## Cleaning of the "pueblo" field

```python
pueblos = menonita_especialistas['pueblo'].unique()
pueblos.sort()
print(pueblos)
```

```python
menonita_especialistas['pueblo'] = menonita_especialistas['pueblo'].str.strip()
```

```python
menonita_especialistas['pueblo'].nunique()
```

## Cleaning of the "nombre" field

```python
menonita_especialistas[['nombre', 'apellidos']] = menonita_especialistas['nombre'].str.split(' ', n=1, expand=True)
```

```python
def split_apellidos(apellidos):
    if apellidos is None:
        return '', ''
    words = apellidos.split()
    if len(words[0]) == 1:
        return ' '.join(words[1:]), words[0]
    else:
        return apellidos, ''
```

```python
menonita_especialistas['apellidos'], menonita_especialistas['inicial'] = zip(*menonita_especialistas['apellidos'].apply(split_apellidos))
```
