## Creating the DataFrame

```python
import pandas as pd
```

```python
data = {
    'nombre': nombre_values,
    'especialidad': especialidad_values,
    'pueblo': pueblo_values,
    'telefono': telefono_values
}
menonita_providers = pd.DataFrame(data)
```

```python
menonita_providers
```

## Cleaning of the "especialidad" field

```python
menonita_providers['especialidad'].value_counts()
```

```python
menonita_providers['especialidad'] = menonita_providers['especialidad'].str.upper()
```

```python
menonita_providers['especialidad'] = menonita_providers['especialidad'].str.replace('PRCTICA GENERAL', 'GENERALISTA').str.replace('PEDIATRA','PEDIATRIA').str.replace('GINECLOGO OBSTETRA','GINECOLOGIA OBSTETRICIA').str.replace('GINECOLOGA','GINECOLOGO').str.replace('GERIATRA','GERIATRIA').str.replace('MEDICINA FAMILIAR','MEDICINA DE FAMILIA')
```

```python
menonita_providers['categoria'] = 'MEDICOS'
```

## Cleaning of the "telefono" field

```python
menonita_providers['telefono'] = menonita_providers['telefono'].str.replace(r'\(|\)', ' ', regex=True).str.replace(' ', '-').str.replace('-787','787').str.replace('-939', '939').str.replace('-NO-DATA', '').str.replace('--','-')
```

## Cleaning of the "pueblos" field

```python
pueblos = menonita_providers['pueblo'].unique()
pueblos.sort()
print(pueblos)
```

```python
menonita_providers['pueblo'] = menonita_providers['pueblo'].str.strip()
```

```python
menonita_providers['pueblo'].nunique()
```

## Cleaning of the "nombre" field

```python
menonita_providers[['nombre', 'apellidos']] = menonita_providers['nombre'].str.split(' ', n=1, expand=True)
```

```python
def split_apellidos(apellidos):
    if apellidos is None:
        return '', ''
    words = apellidos.split()
    if len(words[0]) == 1:
        return ' '.join(words[1:]), words[0]
    else:
        return apellidos, ''
```

```python
menonita_providers['apellidos'], menonita_providers['inicial'] = zip(*menonita_providers['apellidos'].apply(split_apellidos))
```

## Adding new columns for organization

```python
menonita_providers['source_id'] = '002'
```

```python
menonita_providers['tipo_plan'] = 'VITAL'
```

```python
columns = ['source_id','nombre', 'inicial', 'apellidos', 'categoria','especialidad', 'telefono', 'pueblo','tipo_plan']
menonita_providers = menonita_providers[columns]
```
