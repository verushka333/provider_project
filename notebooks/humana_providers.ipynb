## Cleaning of the "nombre" field

```python
def split_initial_end(df, column):
    df['inicial'] = df[column].apply(lambda x: x[-2:] if x[-1].isalpha() and x[-2] == ' ' else '')
    df[column] = df[column].apply(lambda x: x[:-2].strip() if x[-1].isalpha() and x[-2] == ' ' else x.strip())
    return df
```

```python
humana = split_initial_end(humana, 'nombre')
```

```python
humana['inicial'].value_counts()
```

```python
def split_name(df, column):
    df['name'] = df[column].apply(lambda x: ' '.join(x.split()[-2:]) if x.split()[-1] == 'DEL' else x.split()[-1])
    df[column] = df[column].apply(lambda x: ' '.join(x.split()[:-2]) if x.split()[-1] == 'DEL' else ' '.join(x.split()[:-1]))
    return df
```

```python
humana = split_name(humana, 'nombre')
```

```python
humana  = humana.rename(columns={'nombre': 'apellidos'})
```

```python
humana  = humana.rename(columns={'name': 'nombre'})
```

```python
humana  = humana.rename(columns={'dirreccion_2': 'direccion_2'})
```

## Cleaning of the "especialidad" field

```python
humana['especialidad']=humana['especialidad'].str.replace('-',' ')
```

```python
humana['especialidad']=humana['especialidad'].str.replace('OBSTETRICIA GINECOLOGIA','GINECOLOGO OBSTETRA')
```

```python
humana['especialidad']=humana['especialidad'].str.replace('NUTRICIONISTAS','NUTRICIONISTA').str.replace('PERIODONCISTAS','PERIODONCISTA').str.replace('PROSTODONCISTAS','PROSTODONCISTA').str.replace('ORTODONCISTAS','ORTODONCISTA').str.replace('ENDODONCISTAS','ENDODONCIA').str.replace('DENTISTA GENERAL','DENTISTA')
humana['especialidad']=humana['especialidad'].str.replace('PSICOLOGOS','PSICOLOGO').str.replace('PSIQUIATRAS','PSIQUIATRIA').str.replace('GINECOLOGOS','GINECOLOGO')
```
